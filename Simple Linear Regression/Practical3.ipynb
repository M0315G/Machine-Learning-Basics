## Regression

In this notebook I've implemented Regression algorithm. Total 4 versions of it are implemented here:

### Simple Linear Regression

In this part of the notebook I have implemented the Simple Linear Regression with the help of:
- **Gradient Descent** method for optimizing the values of the **```theta```**
- **Normal Equation** method for optimizing the values of the **```theta```**

````python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
````

````python
def load_dataset():
    data = pd.read_csv('data.csv')
    X = np.array(data.iloc[:, 0])
    Y = np.array(data.iloc[:, 1])
    
    split = 0.8
    samples = len(X)
    num = int(split*samples)
    
    X_train1 = np.array(X[:num,])
    X_test1 = np.array(X[num:,])
    Y_train1 = np.array(Y[:num,])
    Y_test1 = np.array(Y[num:,])

    print("The size of X_train is: ",X_train1.shape)
    print("The size of Y_train is: ",Y_train1.shape)
    print("The size of X_test is: ",X_test1.shape)
    print("The size of Y_test is: ",Y_test1.shape)

    return X_train1, Y_train1, X_test1, Y_test1
````

````python
def mse(y_pred, y):
    return np.mean((y - y_pred)**2)
````

````python
def train(X, y, iters=10000, lr = 0.0001):
    # Init parameters
    n_samples = X.shape[0]
    theta1 = 0
    theta0 = 0
    
    for i in range(iters):
        y_pred = X*theta1 + theta0
        
        dtheta1 = (1/n_samples)*np.dot(X.T, (y_pred - y))
        dtheta0 = (1/n_samples)*np.sum((y_pred - y))
        
        theta1 = theta1 - lr*dtheta1
        theta0 = theta0 - lr*dtheta0
        
        if(i%100==0):
            cost = mse(y_pred, y)
            print("Cost after iteration ",i," is ", cost)
            
    return [theta1, theta0]
````

````python
def predict(X, theta1, theta0):
    y_predicted = X*theta1 + theta0
    return y_predicted
````

````python
X_train1, Y_train1, X_test1, Y_test1 = load_dataset()
plt.scatter(X_train1, Y_train1)
plt.show()
````

### By using Gradient Descent

````python
theta1, theta0 = train(X_train1, Y_train1, 500000, 0.0005)
````

````python
# Performance on the Test Set
y_predicted = predict(X_test1, theta1, theta0)
print("MSE on test set is: ",mse(y_predicted, Y_test1))
````

### Multi-variate Linear Regression

In this part of the notebook I have implemented the Simple Linear Regression with the help of:
- **Gradient Descent** method for optimizing the values of the **```theta```**
- **Normal Equation** method for optimizing the values of the **```theta```**

````python
import numpy as np
from sklearn.datasets import load_boston
import pandas as pd
from sklearn.preprocessing import StandardScaler


boston_dataset = load_boston()
````

````python
def load_dataset():
    X1 = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
    Y = boston_dataset.target
    
    split = 0.8
    samples = len(X1)
    num = int(split*samples)
    
    x1 = np.ones((len(X1), ))
    X = np.zeros((samples, X1.shape[1]+1))
    X[:,0] = x1
    X[:,1:] = X1
    
    X_train3 = np.array(X[:num, :])
    X_test3 = np.array(X[num:, :])
    Y_train3 = np.array(Y[:num,])
    Y_test3 = np.array(Y[num:,])

    print("The size of X_train is: ",X_train3.shape)
    print("The size of Y_train is: ",Y_train3.shape)
    print("The size of X_test is: ",X_test3.shape)
    print("The size of Y_test is: ",Y_test3.shape)

    return X_train3, Y_train3, X_test3, Y_test3
````

````python
def normalize_data(data):
    scaler = StandardScaler()
    scaler.fit(data)
    scaler.transform(data)
    
    return data
````

````python
def mse(y_pred, y):
    return np.mean((y - y_pred)**2)
````

````python
def train(X, y, iters=10000, lr = 0.0000001):
    # Init parameters
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)
    
    for i in range(iters):
        y_pred = np.dot(X, weights)
        
        dw = (1/n_samples)*np.dot(X.T, (y_pred - y))
        
        weights = weights - lr*dw
        
        if(i%500==0):
            cost = mse(y_pred, y)
            print("Cost after iteration ",i," is ", cost)
            
    return weights
````

````python
def predict(X, weights):
    y_predicted = np.dot(X, weights)
    return y_predicted
````

### By using Gradient Descent

````python
X_train3, Y_train3, X_test3, Y_test3 = load_dataset()
X_train3 = normalize_data(X_train3)

# Using Gradinet descent to find the optimized theta
weights=train(X_train3, Y_train3, 700000, 0.000005)
````

````python
# Performance on the Test Set
y_predicted = predict(X_test3, weights)
print("MSE on test set is: ",mse(y_predicted, Y_test3))
````

### By using Normal Equation

````python
def normal_eq(X_train3, Y_train3):
    a1 =np.dot(X_train3.T, X_train3)
    a2 = np.dot(X_train3.T, Y_train3)
    theta3 = np.dot(np.linalg.pinv(a1), a2)
    
    return theta3
````

````python
# Using Normal Equation
weights3=normal_eq(X_train3, Y_train3)
y_predicted = predict(X_test3, weights3)
print("MSE on test set is: ",mse(y_predicted, Y_test3))
````

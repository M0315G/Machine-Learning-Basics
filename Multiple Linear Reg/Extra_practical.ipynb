### Using Numpy and other standard Library

```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
def load_dataset():
    X1 = np.array(pd.read_csv("wine.csv"))
    Y = X1[:,11]
    X1 = X1[:,:11]
    #Y = boston_dataset.target
    
    split = 0.8
    samples = len(X1)
    num = int(split*samples)
    
    x1 = np.ones((len(X1), ))
    X = np.zeros((samples, X1.shape[1]+1))
    X[:,0] = x1
    X[:,1:] = X1
    
    X_train = np.array(X[:num, :])
    X_test = np.array(X[num:, :])
    Y_train = np.array(Y[:num,])
    Y_test = np.array(Y[num:,])

    print("The size of X_train is: ",X_train.shape)
    print("The size of Y_train is: ",Y_train.shape)
    print("The size of X_test is: ",X_test.shape)
    print("The size of Y_test is: ",Y_test.shape)

    return X_train, Y_train, X_test, Y_test
```

```python
def normalize_data(data):
    scaler = StandardScaler()
    scaler.fit(data)
    scaler.transform(data)
    
    return data
```

```python
def mse(y_pred, y):
    return np.mean((y - y_pred)**2)
```

```python
def train(X, y, iters=10000, lr = 0.0000001):
    # Init parameters
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)
    
    for i in range(iters):
        y_pred = np.dot(X, weights)
        
        dw = (1/n_samples)*np.dot(X.T, (y_pred - y))
        
        weights = weights - lr*dw
        
        if(i%500==0):
            cost = mse(y_pred, y)
            print("Cost after iteration ",i," is ", cost)
            
    return weights
```

```python
def predict(X, weights):
    y_predicted = np.dot(X, weights)
    return y_predicted
```

```python
X_train1, Y_train1, X_test1, Y_test1 = load_dataset()
X_train1 = normalize_data(X_train1)
```

```python
# Using Gradinet descent to find the optimized theta
weights=train(X_train1, Y_train1, 700000, 0.000005)
```

```python
# Performance on the Test Set
y_predicted = predict(X_test1, weights)
print("MSE on test set is: ",mse(y_predicted, Y_test1))
```

### Using Sklearn Library

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
```

```python
def load_data():
    X1 = np.array(pd.read_csv("wine.csv"))
    Y = X1[:,11]
    X1 = X1[:,:11]
    
    samples = len(X1)
    
    x1 = np.ones((len(X1), ))
    X = np.zeros((samples, X1.shape[1]+1))
    X[:,0] = x1
    X[:,1:] = X1
    
    X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.20)

    print("The size of X_train is: ",X_train.shape)
    print("The size of Y_train is: ",Y_train.shape)
    print("The size of X_test is: ",X_test.shape)
    print("The size of Y_test is: ",Y_test.shape)

    return X_train, Y_train, X_test, Y_test
```

```python
X_train2, Y_train2, X_test2, Y_test2 = load_data()
X_train2 = normalize_data(X_train2)
```

```python
## Using the Built-in regression model

regressor = LinearRegression()
regressor.fit(X_train2, Y_train2)
```

```python
y_pred2 = regressor.predict(X_test2)
print("MSE on test set is: ",mse(y_pred2, Y_test2))
```

## Polynomial Regression

In this notebook, I have implemented the Polynomial Regressiona and compared it with the Simple Linear Regression.

To understand the importance of choosing the right degree of the polynomial, I have implemented 5 different Polynomial Regression with:
- Degree 2
- Degree 3
- Degree 4
- Degree 5
- Degree 10

We'll see as we increase the Degree of the polynomial, the model starts to over fit the data.

It gives the best perfromance with the polynomial of **Degree 3**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression 
```

### Loading the Training Data:

For this notebook, I have used the dataset of Age Vs. Height (https://www.kaggle.com/sakshamjn/heightvsweight-for-linear-polynomial-regression)

```python
dataset = pd.read_csv('HeightVsWeight.csv')
dataset.info()
```

```python
X = dataset.iloc[:,0:1].values
y = dataset.iloc[:,1].values
```

```python
# Function to calculate the MSE of the prediction.
def mse(y_pred, y):
    return np.mean((y - y_pred)**2)
```

### Using Simple Linear Regression

```python
lin_reg1 = LinearRegression()
lin_reg1.fit(X, y)
```

```python
# plotting the result of Simple Linear Regression
fig= plt.figure(figsize=(20,8))
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg1.predict(X), color='blue')
plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Prediction of Simple Linear Regression')
plt.show()
```

```python
y_pred1 = lin_reg1.predict(X)
print("MSE on test set is: ",mse(y_pred1, y))
```

### Using a polynomial of degree 2

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg1 = PolynomialFeatures(degree=2)
X_poly1 = poly_reg1.fit_transform(X)
poly_reg1.fit(X_poly1, y)

lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly1, y)
```

```python
# plotting the result of Simple Linear Regression
fig= plt.figure(figsize=(20,8))
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg2.predict(poly_reg1.fit_transform(X)), color='blue')
plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Prediction by using a Polynomial of Degree 2')
plt.show()
```

```python
y_pred2 = lin_reg2.predict(X_poly1)
print("MSE on test set is: ",mse(y_pred2, y))
```

### Using a Polynomial of Degree 3

```python
poly_reg2 = PolynomialFeatures(degree=3)
X_poly2 = poly_reg2.fit_transform(X)
poly_reg2.fit(X_poly2, y)

lin_reg3 = LinearRegression()
lin_reg3.fit(X_poly2, y)
```

```python
# plotting the result of Simple Linear Regression
fig= plt.figure(figsize=(20,8))
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg3.predict(poly_reg2.fit_transform(X)), color='blue')
plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Prediction by using a Polynomial of Degree 3')
plt.show()
```

```python
y_pred3 = lin_reg3.predict(X_poly2)
print("MSE on test set is: ",mse(y_pred3, y))
```

```python
poly_reg3 = PolynomialFeatures(degree=4)
X_poly3 = poly_reg3.fit_transform(X)
poly_reg3.fit(X_poly3, y)

lin_reg4 = LinearRegression()
lin_reg4.fit(X_poly3, y)
```

- Decision Tree using sklearn with Gini Index as the setting

```python
import numpy as np
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
import graphviz
from sklearn import tree
import pandas as pd
```

#### Function to load data

```python
def load_data():

    data = datasets.load_iris()

    X = data.data
    y = data.target

    print("The total numbe of Samples in the dataset is: ",str(X.shape[0]))

    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)

    return x_train, x_test, y_train, y_test
```

```python
def prediction(x_test, clf_object):

    # Prediction 
    y_pred = clf_object.predict(x_test)
    print("Predicted values: ")
    print(y_pred)

    return y_pred
```

```python
def cal_accuracy(y_test, y_pred):

    print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))

    print("Accuracy : ",accuracy_score(y_test, y_pred)*100)

    print("Report : \n",classification_report(y_test, y_pred))

```

### Part 1:

Decision Tree using Entropy as the setting

```python
def train_using_entropy(x_train, y_train, x_test):

    # Creating a Decision tree classifier
    clf_entropy = DecisionTreeClassifier(criterion="entropy", random_state=100, max_depth=3, min_samples_leaf=5)

    # Training
    clf_entropy.fit(x_train, y_train)

    return clf_entropy
```

```python
x_train, x_test, y_train, y_test = load_data()
clf_entropy = train_using_entropy(x_train, y_train, x_test)
print("\nResults Using Entropy: ")
y_pred_entropy = prediction(x_test, clf_entropy)
cal_accuracy(y_test, y_pred_entropy)
```

```python
tree.plot_tree(clf_entropy)
```

### Part 2:

Decision Tree using Gini Index as the setting

```python
def train_using_gini(x_train, y_train, x_test):

    # Creating a Decision tree classifier
    clf_gini = DecisionTreeClassifier(criterion="gini", random_state=100, max_depth=3, min_samples_leaf=5)

    # Training
    clf_gini.fit(x_train, y_train)

    return clf_gini
```

```python
x_train, x_test, y_train, y_test = load_data()
clf_gini = train_using_gini(x_train, y_train, x_test)
print("\nResults Using Gini Index: ")
y_pred_gini = prediction(x_test, clf_gini)
cal_accuracy(y_test, y_pred_gini)
```

```python
tree.plot_tree(clf_gini)
```
